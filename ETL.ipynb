```python
# *---------------------------------------------- ETL PARTE 2/3 - TRATAMENTO ----------------------------------------------*
```

```python
#Importação de bibliotecas para manipulação
```

```python
import numpy as np
import pandas as pd
```

```python
#Atribuição de uma variável para os dados
```

```python
dados = iris.data
```

```python
#Transformação da variável "dados" para dataframe
```

```python
dados = pd.DataFrame(dados)
dados
```

```python
#Atribuição de uma variável para as classes das flores
```

```python
classes = iris.target
```

```python
#Transformação da variável "classes" para dataframe
```

```python
classes = pd.DataFrame(classes)
classes
```

```python
#Junção dos dataframes "dados" e "classes"
```

```python
df = pd.concat([dados,classes],axis=1)
df
```

```python
#Renomeação de colunas
```

```python
df.columns = ['sepala-comprimento', 'sepala-largura', 'petala-comprimento', 'petala-largura', 'classificacao']
df
```

```python
#Atribuição de uma variável para os nomes das flores
```

```python
nomes = iris.target_names
nomes
```

```python
#Substituição das representações numéricas (0, 1 e 2) para as classes das flores
```

```python
df['classificacao'] = df['classificacao'].replace({0:'setosa', 1:'versicolor', 2:'virginica'})
df
```
